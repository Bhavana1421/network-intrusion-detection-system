```python
random_number = random.randint(0, 22544)
first_row = df.iloc[random_number]
column_names = df.columns

first_row = first_row.to_dict()

result_dict = {col: first_row[col] for col in column_names}
json_data = json.dumps(result_dict)
print(json_data)
```
